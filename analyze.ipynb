```python
exp_name = 'polaris_droid_lang_classifier_step_2k_WITH_pi05_droid_polaris'
import os
from collections import defaultdict
import pandas as pd
```

```python
files = defaultdict(list)
for file in os.listdir(f'runs/{exp_name}'):
    env_name = file.split('_')[0]
    files[env_name].append(file)

results = {}
for env_name, files in files.items():
    lens, succs, progs = [], [], []
    for file in files:
        csv = pd.read_csv(f'runs/{exp_name}/{file}/eval_results.csv')
        lens.extend(csv['episode_length'])
        succs.extend(csv['success'])
        progs.extend(csv['progress'])

    print(env_name, len(lens), len(succs), len(progs))

    results[env_name] = {
        'lens': lens,
        'succs': succs,
        'progs': progs
    }
```

```python
# compute mean and std of the results
import numpy as np
ordered_keys = ['DROID-PanClean', 'DROID-BlockStackKitchen', 'DROID-FoodBussing', 'DROID-MoveLatteCup', 'DROID-OrganizeTools', 'DROID-TapeIntoContainer']
print("~~~~ Best of N=8 Sampling ~~~~~")
for env_name in ordered_keys:
    env_results = results[env_name]
    print(f"-------{env_name}-------, Rollouts = {len(env_results['lens'])}")
    # print(f"Episode Lengths: Mean={np.mean(env_results['lens'])  } , Std={np.std(env_results['lens'])  } ")
    print(f"Success Rate: Mean={np.mean(env_results['succs']) * 100 } %, Std={np.std(env_results['succs']) * 100 } %")
    print(f"Progress: Mean={np.mean(env_results['progs']) * 100 } %, Std={np.std(env_results['progs']) * 100 } %")
```
